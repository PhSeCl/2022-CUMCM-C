# 数据勘察

通过项目包读取已验证的 CSV，检查形状、字段、缺失值和成分总和。本 Notebook 不修改源数据。

```python
from pathlib import Path

from cumcm2022c.data.load import load_table
from cumcm2022c.data.validate import validate_composition_sums

DATA_DIR = Path("../data/raw") if Path.cwd().name == "notebooks" else Path("data/raw")
```

```python
tables = {name: load_table(name, DATA_DIR) for name in ("form_1", "form_2", "form_3")}
{name: {"shape": frame.shape, "columns": frame.columns.tolist()} for name, frame in tables.items()}
```

```python
missing_summary = {name: frame.isna().sum().to_dict() for name, frame in tables.items()}
missing_summary
```

```python
form_2_components = tables["form_2"].columns[1:].tolist()
form_3_components = tables["form_3"].columns[2:].tolist()
validated = {
    "form_2": validate_composition_sums(tables["form_2"], form_2_components),
    "form_3": validate_composition_sums(tables["form_3"], form_3_components),
}
{
    name: frame[["composition_sum", "is_valid_composition"]].describe(include="all")
    for name, frame in validated.items()
}
```
